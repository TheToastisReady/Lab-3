```python
import statistics as st
import matplotlib.patches as pch
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import scipy.stats as ss
% matplotlib inline

PVC = open("PoliceViolence_city.csv", "r", errors="ignore");
KBP = open("KilledByPolice.csv", "r", errors="ignore");

cur = [];
cur2 = [];

for aa in PVC:
    cur.append(aa);
for ab in KBP:
    cur2.append(ab);

del cur2[0];
del cur[0];

for ba in range(len(cur)):
    cur[ba] = cur[ba].split(",");
for bb in range(len(cur2)):
    cur2[bb] = cur2[bb].split(",");


#VCR = violent crime rate
vcr = [];
vcrc = [];
bcrc = [];
peopler = [];
phrblack = [];
    

for d in range(len(cur)):
    vcr.append(float(cur[d][17]));
    vcrc.append(float(cur[d][19]));
    peopler.append((float(cur[d][6])/float(cur[d][5]))*100);
    phrblack.append(float(cur[d][21]));

reg1 = ss.linregress(vcr, vcrc);

print(reg1[2], reg1[0]);
for e in range(len(vcr)):
    colr = np.random.rand(3);
    graphy = plt.scatter(vcr[e], vcrc[e], color=colr);
    
```

```python
colr = np.random.rand(3);
reg2 = ss.linregress(peopler, phrblack);

print(reg2[2], reg2[0]);

for f in range(len(peopler)):
    plt.scatter(peopler[f], phrblack[f], color = colr);
    
```

These two graphs show the r values of violent crime rate of each state, and the police homocide rate for each state compared to the total population. What we can see from the second r value that the more black people there are in a population, the less likely they are to be killed by police. What I did in this code was create scatter plots based off the police homicide rate versus total population, and police homicide rate on black people versus the black population of the area. What the correlation coefficients show us for both graphs is that as the black population increases, there is less police homicide of black people, and that as the total population increases, the total police homicide rate increases.
